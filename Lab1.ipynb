### Ćwiczenie 3 

Wyrysowanie wykresów funkcji  na podanych przedziałach:

    x>-1 oraz x<1
    x>-6 oraz x<6
    x>0 oraz x<5
przy pomocy biblioteki matplotlib.

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
fun=lambda x: x**2+5  #deklaracja funkcji
```

```python
#definicja przedziałow w którym dokonujemy wyrysowania
x1 =np.arange(-1,1,0.1)
x2 = np.arange(-6,6,0.1)
x3 = np.arange(0,6,0.1)
```

```python
#funkcja na przedziale <-1,1>
plt.plot(x1,fun(x1),label="y = x**2+5")
plt.title("Wykres funkcji x**2 + 5 na przedziale <-1,1> ")
plt.xlabel("Argumety funkcji")
plt.ylabel("Wartości funkcji")
plt.legend()
plt.show()
```

```python
#funkcja na przedziale <-6,6>
plt.plot(x2,fun(x2),label="y = x**2+5")
plt.title("Wykres funkcji x**2 + 5 na przedziale <-6,6> ")
plt.xlabel("Argumety funkcji")
plt.ylabel("Wartości funkcji")
plt.legend()
plt.show()
```

```python
#funkcja na przedziale <0,5>
plt.plot(x3,fun(x3),label="y = x**2+5")
plt.title("Wykres funkcji x**2 + 5 na przedziale <0,5> ")
plt.xlabel("Argumety funkcji")
plt.ylabel("Wartości funkcji")
plt.legend()
plt.show()
```
